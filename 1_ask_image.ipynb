```python
from vertexai.preview.vision_models import ImageQnAModel
from vertexai.preview.vision_models import Image
```

```python
image="various_food.jpg"
```

```python
def show_image(image):
    cloud_next_image = Image.load_from_file(image)
    cloud_next_image.show()
```

```python
show_image(image)
```

```python
# Ask a question about the image
def ask_image(question,number_of_results):
    
    image_qna_model = ImageQnAModel.from_pretrained("imagetext@001")
    cloud_next_image = Image.load_from_file(image)
    
    return image_qna_model.ask_question(
        image=cloud_next_image, question=question, number_of_results=number_of_results)
    
```

```python
ask_image("What is in the image?",number_of_results=3)
```

```python
ask_image("Is the food good quality or perished?",number_of_results=3)
```

```python
ask_image("What types of fruit are there?", number_of_results=3)
```

```python
ask_image("Are eggs in the image?",number_of_results=1)
```

```python
ask_image("Is fish in the image?",number_of_results=1)
```

```python
ask_image("What fish is in the image?",number_of_results=3)
```

```python
ask_image("What meat is in the image?",number_of_results=3)
```

```python
ask_image("How many items are in the image?",number_of_results=1)
```

```python
ask_image("Can you assess the weight of all the items in the image?",number_of_results=1)
```

```python
ask_image("What colors do you see in the image?",number_of_results=3)
```
